```python
#Building a model on Bluffing detector of a Candidate stating his Package
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
#Declaration of Independant & Dependant variables, here we are not considering the title as Independant variable as it is not required
X = dataset.iloc[:,1:2].values# ,1:2 is to declare that X is a matrix and not a vector
y = dataset.iloc[:,2].values
```

```python
#Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
#Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)# 1st column is filled with one in order to have the value of the constant b0
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
#Visualizing the linear regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X),color ='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
# Red Points - Observation points
# Blue line - Prediction
```

```python
#Visualizing Polynomial Regression results
#Visualizing the linear regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)),color ='blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
# Red Points - Observation points
# Blue line - Prediction
```
